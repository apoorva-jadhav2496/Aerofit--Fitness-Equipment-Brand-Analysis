# **Univariate Analysis for the quantitative attributes**


```python
# Histogram and kde plot for quantitative columns to understand their distribution 
plt.figure(figsize=(18,18))
plt.subplot(3, 3, 1)
sns.histplot(data = df, x = "Age" , kde=True,  color = "orange")
plt.title("Distribution of Age" ,fontstyle='italic')

plt.subplot(3, 3, 2)
sns.histplot(data = df, x = "Education" , kde=True, color = "orange")
plt.title("Distribution of Education" ,fontstyle='italic')

plt.subplot(3, 3, 3)
sns.histplot(data = df, x = "Usage" , kde=True, color = "orange")
plt.title("Usage per week" ,fontstyle='italic')

plt.subplot(3, 3, 4)
sns.histplot(data = df, x = "Fitness" , kde=True, color = "orange")
plt.title("Distribution of Fitness" ,fontstyle='italic')

plt.subplot(3, 3, 5)
sns.histplot(data = df, x = "Income" ,  kde=True, color = "orange")
plt.title("Distribution of Income" ,fontstyle='italic')

plt.subplot(3, 3, 6)
sns.histplot(data = df, x = "Miles" ,  kde=True ,color = "orange")
plt.title("Average no of Miles " ,fontstyle='italic')
plt.show
```

**Detecting Outliers using Box Plot**

```python
plt.figure(figsize=(12,12))
plt.subplot(3, 3, 1)
sns.boxplot(data = df, x = "Age" , color = "orange")

plt.subplot(3, 3, 2)
sns.boxplot(data = df, x = "Education" , color = "orange")

plt.subplot(3, 3, 3)
sns.boxplot(data = df, x = "Usage" , color = "orange")

plt.subplot(3, 3, 4)
sns.boxplot(data = df, x = "Fitness" , color = "orange")

plt.subplot(3, 3, 5)
sns.boxplot(data = df, x = "Income" , color = "orange")

plt.subplot(3, 3, 6)
sns.boxplot(data = df, x = "Miles" , color = "orange")
plt.show
```

# **Bivariate Analysis**
**Check if features like marital status, age have any effect on the product purchased**

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 3, 1)
sns.countplot(data = df, x = "Product", hue = "Gender" ,edgecolor="0.08" )
plt.title("Product vs Gender" , fontstyle ="italic")

plt.subplot(1, 3, 2)
sns.countplot(data = df, x = "Product", hue = "MaritalStatus", edgecolor="0.08")
plt.title("Product vs Marital Status" , fontstyle ="italic")

plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='Product', y="Age")
plt.title("Product vs Age" , fontstyle ="italic")
```

Equal number of males and females have purchased KP281 product which shows many of the customers have just started their fitness journey as it is entry level treadmill.

Almost same amoun of males and female have purchased the product KP481.

Most of the Male customers have purchased the KP781 product which is costliest product of Aerofit.

Customer who is Partnered, is more likely to purchase the product.

Customers purchasing products KP281 & KP481 are having same Age median value and Customers whose age lies between 25-30, are more likely to buy KP781 product.

 **Representing the marginal probability like - what percent of customers have purchased KP281, KP481, or KP781 in a table using cross-tab**

```python
df2 = pd.crosstab(index=df['Product'],
            columns=df["Product"],
            margins=True)

df3 = df2*100/180
df3
```
